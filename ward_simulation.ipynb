```python
# Create random three 2D Gaussian distrobutions
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

m=[[0.35,0.4],[0.65,0.4],[0.5,0.65]]
s=[[0.01,0],[0,0.01]];

# Define dense grid
Nx=100
Ny=100
X=np.meshgrid(np.linspace(0,1,Nx),np.linspace(0,1,Ny))
Xval=np.vstack((X[0].flatten(),X[1].flatten())).T
Y1=2*multivariate_normal.pdf(Xval,m[0],s).T
Y2=1.5*multivariate_normal.pdf(Xval,m[1],s).T
Y3=multivariate_normal.pdf(Xval,m[2],s).T
Y=np.column_stack((Y1,Y2,Y3))

Yimg=np.reshape(Y1+Y2+Y3,[Nx,Ny])
noise_level=10
noise=ndimage.gaussian_filter(
        np.random.uniform(0, noise_level,size=[Nx,Ny]),sigma=1)

f=plt.figure()
f.add_subplot(121)
plt.imshow(Yimg)
plt.axis('off')
f.add_subplot(122)
plt.imshow(Yimg+noise)
plt.axis('off')
plt.show()

f=plt.figure()
f.add_subplot(131)
plt.imshow(np.reshape(Y1,[Nx,Ny])+noise)
plt.axis('off')
f.add_subplot(132)
plt.imshow(np.reshape(Y2,[Nx,Nx])+noise)
plt.axis('off')
f.add_subplot(133)
plt.imshow(np.reshape(Y3,[Nx,Ny])+noise)
plt.axis('off')
plt.show()
```

```python
# Build connectivity matrix

ind=np.reshape(np.arange(0,Nx*Ny),[Nx,Ny])
A=np.zeros((Ny*Ny,Nx*Ny),dtype=int)
for x in np.arange(0,Nx):
    for y in np.arange(0,Ny):
        sub=np.asarray([[x-1,y-1],[x-1,y],[x-1,y+1],[x,y-1],
                        [x,y+1],[x+1,y-1],[x+1,y],[x+1,y+1]])
        sub=sub[np.logical_not(np.logical_or(np.logical_or(np.logical_or(sub[:,0]<0,
                        sub[:,1]<0),sub[:,0]>=Nx),sub[:,1]>=Ny))]
        A[ind[x,y],ind[sub[:,0],sub[:,1]]]=1
```

```python
# Perform Ward clustering on BPnd data

from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster.hierarchical import _hc_cut # Internal function to cut ward tree, helps speed up things a lot
import sklearn.metrics.cluster as metrics

N_iter=1
K_range=np.arange(2,20)
K_range=[4]
noise_level=10

ars=np.empty([N_iter,len(K_range)])
ami=np.empty([N_iter,len(K_range)])
mdl1=AgglomerativeClustering(compute_full_tree=True, connectivity=A,linkage='ward')
mdl2=AgglomerativeClustering(compute_full_tree=True, connectivity=A,linkage='ward')

for nk in K_range:
    for ni in np.arange(0,N_iter):
        noise1=np.reshape(ndimage.gaussian_filter(
                np.random.uniform(0, noise_level,size=[Nx,Ny]),sigma=1),[Nx*Ny])
        noise2=np.reshape(ndimage.gaussian_filter(
                np.random.uniform(0, noise_level,size=[Nx,Ny]),sigma=1),[Nx*Ny])
        mdl1.fit(Y+noise1[:,np.newaxis])
        mdl2.fit(Y+noise2[:,np.newaxis])

        for K,nk in zip(K_range,np.arange(0,len(K_range))):
            # Cut trees
            # NOTE: labels start at 0, so add one to distinguish from medial wall
            labels1=_hc_cut(K,mdl1.children_,mdl1.n_leaves_)+1
            labels2=_hc_cut(K,mdl2.children_,mdl2.n_leaves_)+1

            # Compute metrics
            ars[ni,nk]=metrics.adjusted_rand_score(labels1, labels2)
            ami[ni,nk]=metrics.adjusted_mutual_info_score(labels1, labels2)

# Plot metrics
plt.figure(figsize=(8,2))
plt.axis([0,1,K_range[0],K_range[-1]])
plt.subplot(1,2,1)
plt.errorbar(K_range,ars.mean(axis=0),yerr=ars.std(axis=0))
plt.title('Adjusted Rand Index')
plt.subplot(1,2,2)
plt.errorbar(K_range,ami.mean(axis=0),yerr=ami.std(axis=0))
plt.title('Adjusted Mutual Information')
plt.show()
```
